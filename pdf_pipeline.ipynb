# INSERT PDF INTO THE TABLE1

```python
import psycopg2
import uuid
import os
```

```python
connection=psycopg2.connect(
    host="localhost",
    database="pdf_pipeline",
    user="postgres",
    port="5432",
    password="root"
)
cursor=connection.cursor()
print("Database connection successful")
```

```python
for file in os.listdir("sample_pdfs"):
    filepath=os.path.join("sample_pdfs",file)
    with open(filepath,"rb") as f:
        file_data=f.read()
        file_id=str(uuid.uuid4())
        insert_query="""INSERT  INTO t1(id,blob_object) VALUES(%s,%s)"""
        data_to_insert=(file_id,file_data)
        cursor.execute(insert_query,data_to_insert)
connection.commit()
print("All files inserted successfully")
```

```python
cursor.close()
connection.close()
```
